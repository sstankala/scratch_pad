```python
full_dot = '●'
empty_dot = '○'

def create_character(cname, strength, intelligence, charisma):
    if not (isinstance (cname,str)):
        return f'The character name should be a string'
    elif(len(cname) >10):
          return f'The character name is too long'
    elif ' ' in cname:
           return f'The character name should not contain spaces'

    if not (isinstance (strength, int) and isinstance (intelligence,int) and isinstance (charisma, int)):
        return f'All stats should be integers'
    elif  (strength <1 or intelligence <1  or charisma < 1):
        return f'All stats should be no less than 1'
    elif (strength > 4 or intelligence >4 or charisma > 4):
        return f'All stats should be no more than 4'
    elif not (strength+intelligence+charisma) == 7:
        return f'The character should start with 7 points'
    else:
      result = f"{cname}\n"
      result += f"STR {full_dot * strength}{empty_dot * (10 - strength)}\n"
      result += f"INT {full_dot * intelligence}{empty_dot * (10 - intelligence)}\n"
      result += f"CHA {full_dot * charisma}{empty_dot * (10 - charisma)}"
    return result

print(create_character("ren", 4, 2, 1))
```
