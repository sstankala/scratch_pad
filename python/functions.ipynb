```python
def hello(count):
    if count == 0:
        return
    print('Hello World')
    hello(count - 1)

hello(3)
```

```python
# custom function with return
def calculate_sum(a, b):
    return (a + b)

my_sum = calculate_sum(3, 1)
print(my_sum)
```
