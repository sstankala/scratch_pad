```python
my_str = 'hello world'

uppercase_my_str = my_str.upper()
print(uppercase_my_str)

lowercase_my_str = uppercase_my_str.lower()
print('Upper case string:', uppercase_my_str)
print('Lower case string:', lowercase_my_str)

```

```python
my_str = 'hello python world'

replaced_my_str = my_str.replace('hello', 'hi')
print(replaced_my_str)
```

```python
my_str = 'hello python lovers'

split_words = my_str.split()
print(split_words)
```

```python
my_list = ['hello', 'world']

joined_my_str = ' Test join with '.join(my_list)
print(joined_my_str)
```

```python
my_str = 'Test hello world'

starts_with_test = my_str.startswith('Test')
print(starts_with_test)

ends_with_war = my_str.endswith("war")
print(ends_with_war)
```

```python
my_str = 'hello python world'

world_index = my_str.find('python')
print(world_index)
```

```python
my_str = 'hello python world, why do you love python so much?'

count_my_python = my_str.count('python')
print(count_my_python)
str_replace_string = my_str.replace('python','python3')
print(str_replace_string)
```
