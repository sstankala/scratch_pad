```python
# augment with addition
my_var = 100
my_var += 50
print(my_var)

# augment with substraction
my_var = 100
my_var -=75
print(my_var)
```

```python
#augment with multiplication
my_var = 10
my_var *=5
print(my_var)

#augment wit division
my_var = 50
my_var /=10
print(my_var)
```

```python
#augment with mod
my_var = 50
my_var %=9
print(my_var)

#augment with floor division
my_var = 50
my_var //=7
print(my_var)

#augment with power
my_var = 100
my_var **=3
print(my_var)
```
